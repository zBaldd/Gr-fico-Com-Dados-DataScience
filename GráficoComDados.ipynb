```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv_file_path = 'Population.csv'
populacao_data = pd.read_csv(csv_file_path)
```

```python
print(populacao_data.head())
```

```python
# Selecionar dados para um país específico (por exemplo, Brasil)
pais = 'Brazil'
populacao_pais = populacao_data.loc[populacao_data['Country Name'] == pais]

# Plotar gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(populacao_pais.columns[1:], populacao_pais.values[0][1:])
plt.title(f'População Total de {pais}')
plt.xlabel('Ano')
plt.ylabel('População Total')
plt.grid(True)
plt.show()
```

```python
# Selecionar dados de população do Brasil
populacao_brasil = populacao_data.loc[populacao_data['Country Name'] == 'Brazil'].iloc[:, 1:].values.flatten()

# Converter os dados para float
populacao_brasil_float = [float(x) for x in populacao_brasil if str(x).replace('.', '').isdigit()]

# Plotar histograma
plt.figure(figsize=(10, 6))
plt.hist(populacao_brasil_float, bins=20, edgecolor='black')
plt.title('Distribuição da População Total do Brasil ao Longo dos Anos')
plt.xlabel('População Total')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()





```

```python
anos = populacao_pais.columns[4:]  # Começando da quinta coluna, onde estão os anos
populacao_anos = populacao_pais.iloc[:, 4:]

# Tratamento de valores ausentes e conversão para float
populacao_anos_numeric = populacao_anos.apply(pd.to_numeric, errors='coerce').fillna(0)

# Verificando o comprimento dos rótulos e das proporções
print("Número de rótulos:", len(anos))
print("Número de proporções:", len(populacao_anos_numeric))

# Plotando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(populacao_anos_numeric.sum(), labels=anos, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição da População por Ano no Brasil')
plt.axis('equal')
plt.show()
```
